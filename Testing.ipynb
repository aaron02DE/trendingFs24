```python
import pickle
import numpy as np
import os
import sys
import matplotlib.pyplot as plt

y_data = np.load('y.pkl', allow_pickle=True)
metadata = np.load('metadata.pkl', allow_pickle=True)
x_data = np.load('x.pkl', allow_pickle=True)
```

```python
print(len(metadata))
```

```python
#Get Measurements in Dictionary

Measurements = {}

for Id in range(41385):
    if (metadata[Id]["trigger_time"])[:10] != (metadata[Id-1]["trigger_time"])[:10]:                 
        DeviceId = (metadata[Id]["device_id"])
        ElapsedTime = (metadata[Id]["elapsed_time_h"])

        if DeviceId not in Measurements:
            Measurements[DeviceId] = {}             
                                                                 
        Measurements[DeviceId][ElapsedTime] = {                                                             
                                "IdStart"     : Id,
                                "IdEnd"       : Id+92}
        
    else:
        continue

myKeys = list(Measurements.keys())
myKeys.sort()
sorted_Measurements = {i: Measurements[i] for i in myKeys}
 
print(sorted_Measurements)

"""
Dictionary Measurements: {DeviceID: {ElapsedTime: {IdStart: IdStart, IdEnd: IdEnd}}}
"""
```

```python
#Get Measurements for Device xy

Beschleunigung_Horizontal_58    = 0
Beschleunigung_Vertikal_60      = 1
CAN_Flow                        = 2
Druck_DS_P2                     = 3
Druck_SS_P1                     = 4
Strom                           = 5

DeviceId = 31427

NumberOfMeasurements = (len(Measurements[DeviceId]))
for n in range(NumberOfMeasurements):
    ElapsedTime = n * 500
    IdStart = Measurements[DeviceId][ElapsedTime]["IdStart"]
    IdEnd   = Measurements[DeviceId][ElapsedTime]["IdEnd"] 

    x = range(0,93,1)
    y = []
    for IdStart in range(IdStart,IdEnd+1):
        y.append(np.mean(x_data[IdStart][CAN_Flow]))
    plt.plot(x, y, label=ElapsedTime)

# naming the x axis
plt.xlabel('Time')
# naming the y axis
plt.ylabel('CAN_Flow')
 
# giving a title to my graph
plt.title("Graph for Sensor: "+str(DeviceId))
 
# function to show the plot
plt.show()
```
